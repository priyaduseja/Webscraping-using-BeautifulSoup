```python
from bs4 import BeautifulSoup
import requests
```

```python
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",}
```

```python
r = requests.get("https://www.imdb.com/chart/top/",headers=headers, verify=False)
html = r.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
body = soup.find('ul',{'class':'ipc-metadata-list ipc-metadata-list--dividers-between sc-a1e81754-0 eBRbsI compact-list-view ipc-metadata-list--base'})
rows = body.findAll('li')
for row in rows:
    list_of_movies = row.find('div',{'class':'ipc-metadata-list-summary-item__c'})
    for li in list_of_movies:
            year = li.find('span',{'class':'sc-b0691f29-8 ilsLEX cli-title-metadata-item'})
            url = list_of_movies.a['href']
            movieurl = f"https://www.imdb.com{url}"
            
            r2 = requests.get(movieurl,verify=False, headers=headers)
            html = r2.text
            soup2 = BeautifulSoup(html, 'html.parser')
            
            infoul = soup2.find('ul',{'class':'ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt'})
            li = infoul.findAll('li')
            if len(li) >=3:
                movieDuration = li[2].text.strip()
            else:
                movieDuration = "N/A"
            
            genresDiv = soup2.find('div',{'data-testid':'genres'})
            a = genresDiv.findAll('a')
            genres = []
            for genre in a:
                genres.append(genre.text.strip())
            genres = ','.join(genres)
            print(list_of_movies.h3.string, year.string, movieDuration, genres)
            print('------------')
```
